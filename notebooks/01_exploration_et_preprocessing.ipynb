```python
def clean_value(valeur):
    v = valeur.strip()
    valeurs_invalides = ['??', '#BOŞ!', '#BOÞ!', '']
    if v in valeurs_invalides:
        return "N/A"
    v_modifiee = v.replace(',', '.')
    try:
        float(v_modifiee)
        return v_modifiee
    except ValueError:
        return v
```

```python
donnees_nettoyees = []

with open('../data/data.csv', 'r', encoding='latin-1') as fin:
    next(fin)
    for ligne in fin:
        token = ligne.strip().split(';')

        try:
            if len(token) >= 20:
                ligne_propre = [
                    clean_value(token[1]),  # Sex
                    clean_value(token[2]),  # Age
                    clean_value(token[4]),  # Arrival mode
                    clean_value(token[5]),  # Injury
                    clean_value(token[7]),  # Mental
                    clean_value(token[8]),  # Pain
                    clean_value(token[9]),  # NRS_pain
                    clean_value(token[10]), # SBP
                    clean_value(token[11]), # DBP
                    clean_value(token[12]), # HR
                    clean_value(token[13]), # RR
                    clean_value(token[14]), # BT
                    clean_value(token[15]), # Saturation
                    clean_value(token[19])  # KTAS_expert
                ]
                donnees_nettoyees.append(ligne_propre)

        except IndexError:
            continue

print(f"Tableau créé avec {len(donnees_nettoyees)} lignes.")
```

```python
# Définir les noms des colonnes que tu as extraites
colonnes = [
    'Sex', 'Age', 'Arrival_mode', 'Injury', 'Mental', 'Pain', 
    'NRS_pain', 'SBP', 'DBP', 'HR', 'RR', 'BT', 'Saturation', 'KTAS_expert'
]
```

```python
# Créer le nouveau DataFrame
df_final = pd.DataFrame(donnees_nettoyees, columns=colonnes)
```

```python
# Remplacer tes "N/A" par de vrais NaN pour que Pandas les reconnaisse
df_final = df_final.replace('N/A', np.nan)

# Afficher un aperçu
df_final.head()
```

```python
# Convertir les colonnes en numérique (force les erreurs en NaN)
cols_numeriques = ['Age', 'SBP', 'DBP', 'HR', 'RR', 'BT', 'Saturation']

for col in cols_numeriques:
    df_final[col] = pd.to_numeric(df_final[col], errors='coerce')

# Vérifier les types
print(df_final.dtypes)
```
